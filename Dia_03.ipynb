```
import pandas as pd
```

```
dados = pd.read_csv('sds_fifa.csv')
```

```
dados
```

```
# Pega os primeiros dados
dados.head()
```

```
# Pega os últimos dados
dados.tail()
```

```
# Pegar amostra aleatória
dados.sample(10)
```

```
# Tirar informações gerais dos dados
dados.info()
```

```
# Tirar informações numéricas (Estatísticas)
dados.describe().T
```

```
dados['Salario'] = dados['Salario'].str.replace('€', '')
```

```
dados.sample(5)
```

```
# Eliminar dados vazios
dados = dados.dropna()
```

```
dados.info()
```

```
# Obter valores únicos
dados['Salario'].unique()
```

```
# Obter quantidade de valores únicos
dados['Salario'].nunique()
```

```
dados['Salario'] = dados['Salario'].str.replace('K', '000')
```

```
dados
```

```
dados.info()
```

```
dados['Salario'] = dados['Salario'].astype(str).astype(int)
```

```
dados.info()
```

```
dados.describe().T
```

```
dados.drop_duplicates(inplace=True)
```

```
dados.info()
```
